```python
print("Sistema de calculo de áreas geometricas de Diego Bátiz")
print("1.Cuadrado")
print("2.Rectángulo")
print("3.Triangulo")
print("4.Circulo")

area = int(input("Que figura quieres calcular? "))

if area == 1:
  lado = float(input("Ingresa el lado:"))
  resultado = lado ** 2
  print("El area del cuadrado es:", resultado)


if area == 2:
  base = float(input("Ingresa la base:"))
  altura = float(input("Ingresa la altura"))
  resultado = base * altura
  print("El area del rectangulo es:" , resultado)

if area == 3:
  base = float(input("Ingresa la base:"))
  altura = float(input("Ingresa la altura"))
  resultado = base * altura / 2
  print("El area del triangulo es:" , resultado)

if area == 4:
  radio = float(input("Ingresa el radio:"))
  resultado = 3.1416 * radio ** 2
  print("El area del circulo es:" , resultado)



```
