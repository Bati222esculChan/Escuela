```python
articulos = []

while True:
    print("1. Crear artículo")
    print("2. Ver artículos")
    print("3. Editar artículo")
    print("4. Borrar artículo")
    print("5. Salir")

    opcion = input("Elige una opción: ")

    if opcion == "1":
        nombre = input("Nombre: ")
        precio = float(input("Precio: "))
        cantidad = int(input("Cantidad: "))

        articulos.append([nombre, precio, cantidad])

    elif opcion == "2":
        for articulo in articulos:
            print(articulo)

    elif opcion == "3":
        nombre = input("Nombre del artículo a editar: ")

        for articulo in articulos:
            if articulo[0] == nombre:
                articulo[0] = input("Nuevo nombre: ")
                articulo[1] = float(input("Nuevo precio: "))
                articulo[2] = int(input("Nueva cantidad: "))
                print("Artículo editado.")
                break

    elif opcion == "4":
        nombre = input("Nombre del artículo a borrar: ")

        for articulo in articulos:
            if articulo[0] == nombre:
                articulos.remove(articulo)
                print("Artículo borrado.")
                break

    elif opcion == "5":
      break

    else:
        print("Opción no válida.")
```
